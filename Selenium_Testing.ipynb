```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
import time



path= "c:\Program Files\chromedriver-win64\chromedriver.exe" 
driver=webdriver.Chrome(service=Service(path)) 
driver.get("http://127.0.0.1:8000/") 


# Wait for the username input to be present
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "username"))
)


# Wait for the login form to load
wait = WebDriverWait(driver, 4)



#Test 1: Testing the login page and move to home page
username_field = driver.find_element(By.XPATH, '//*[@id="username"]') 
password_field = driver.find_element(By.XPATH, '//*[@id="password"]')  

username_field.send_keys("habibah")  
password_field.send_keys("1")  

#Locating the login button and submit the form
login_button = driver.find_element(By.XPATH, '//html/body/div[1]/form/button') 
login_button.click() 


time.sleep(3) 

# Verifying that we are in the homepage 
welcome_message = driver.find_element(By.XPATH, "//*[contains(text(), 'Welcome')]")  # Adjust based on the homepage content

if welcome_message:
    print("Login successful and homepage loaded successfully!")
else:
    print("Login failed or homepage did not load.")




# Test 2 Verify the presence of the navbar and "VistaNest" logo
navbar = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "navbar")))
logo = driver.find_element(By.CLASS_NAME, "navbar-brand")

assert "VistaNest" in logo.text, "Logo text is incorrect!"




# Test 3: Test the "Search" link in the navbar
search_link = driver.find_element(By.LINK_TEXT, "Search")
assert search_link.is_displayed(), "Search link is not displayed!"
search_link.click()
time.sleep(1)



# Test 4: Test the "Search" by searching a specified thing
search_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "id_query")) 
    )


search_query = "Green Road"
search_input.send_keys(search_query)
search_button = driver.find_element(By.XPATH, "//button[@type='submit']")
search_button.click()


# Wait for results or the "no results" message
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "search-result-card"))
    )
    print(f"Search performed successfully with query: {search_query}. Results were found.")
except TimeoutException:
        print(f"Search performed successfully with query: {search_query}, but no results were found.")

time.sleep(7)
driver.close()
```
